```python
show(html("Функция смещения квадрата"))
def shift_matrix(matrix_, shift_x, shift_y):
    return matrix_ + matrix([ [shift_x, shift_y] for i in range(matrix_.nrows())])
```

```python
show(html("Функция поворота квадрата"))
def rotate_matrix(angle):
    phi = math.radians(angle)
    return matrix([
        [cos(phi), -sin(phi)],
        [sin(phi), cos(phi)]])
```

```python
h = 2
w = 2
rect = matrix([[0, 0], [0, h], [w, h], [w, 0], [0, 0]])

new_rect = []
new_rect = rect * rotate_matrix(-60)
new_rect = shift_matrix(new_rect, 3, 1)

rect = shift_matrix(rect, 1, 0)

res = line(new_rect, color = 'red')
res = res + line(rect, color = 'green')

show(res, aspect_ratio = 1)
```
