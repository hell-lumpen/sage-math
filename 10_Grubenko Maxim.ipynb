```python
x = var("x")

f = (x**2 + 3*cos(1/2*x**2) - 1) / (3*x) - 1/3

show(LatexExpr(f"f(x) = {latex(f)}"))
```

```python
show(plot(f, 1, 3))
```

```python
def newton(f, x_0):
    df = f.derivative()
    
    x1 = x_0 - f(x = x_0) / df(x = x_0) # https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D1%82%D0%BE%D0%B4_%D0%9D%D1%8C%D1%8E%D1%82%D0%BE%D0%BD%D0%B0#:~:text=%D0%92%20%D1%8D%D1%82%D0%BE%D0%BC%20%D1%81%D0%BB%D1%83%D1%87%D0%B0%D0%B5,%D1%81%D0%B2%D0%BE%D0%B4%D0%B8%D1%82%D1%81%D1%8F
    x2 = x1 - f(x = x1) / df(x = x1)
    
    l = df(x = x1)*(x - x1) + f(x = x1) # график функции касательной к графику в точки x1
    
    while abs(x2.n(digits=10) - x1.n(digits=10)) > 0.0000000001:
        x1 = x2 - f(x = x2) / df(x = x2)
        x2 = x1

    return x2, l
```

```python
x1, l = newton(f, 2)
x1 = x1.n(digits=10)
x2, l1 = newton(f, 2.5)
x2 = x2.n(digits=10)
show("x1 = ", x1)
show("x2 = ", x2)
grapfic = plot(f, 1, 3)
grapfic += plot(l, 1, 3)
grapfic += plot(l1, 1, 3)
grapfic += point((x1, 0), size = 30)
grapfic += point((x2, 0), size = 30)
grapfic.show(xmin = 1, xmax = 3, ymin = -0.5, ymax = 0.5)
```
