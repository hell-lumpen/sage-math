```python
var("x")
y(x) = sin(2*x**3)**2/x**3 + x

show(plot(y, -2, 2))
```

<h2>Область определения функции</h2>

<br>Область определения: $x \neq 0$, т.к. функция содержит x<sup>3</sup> в знаменателе дроби</br>
    
<h2>Проверка функции на четность, нечетность и периодичность</h2>
<br>Функция четная, когда у(-х) = у(х)</br>
<br>Функция нечетная, когда у(-х) = -у(х)</br>
<br>Функция периодичная, когда у(х) = у(х+Т) = у(х-Т), где Т - период функции</br>

```python
if (y(-x) == y(x)):
    print("функция четная")
if (y(-x) == -y(x)):
    print("функция нечетная")
```

По графику видно, что функция не является периодической, т.к. график функции - возрастает

<h2>Точки пересечения графика с осями координат</h2>

По области определения функции $x \neq 0$ значит функция никогда не пересекает ось ординат<br>

```python
sols = solve(y==0, x)
show(sols)
```

Из решения и графика функции видим, что ось абсцисс также никогда не пересекает

```python
d_f = y.diff(x)
graf = plot(d_f, -2, 2)
graf += point((find_root(d_f, -1, -0.5), 0), size = 50, color = 'red')
show(html("График производной функции"))
graf.show(aspect_ratio=0.2) 
show(html("В случае этой точки - это локальный минимум. Глобальных максимумов и минимумов - нет"))
```

<h2>Непрерывность. Наличие точек разрыва и их классификация</h2>

```python
lr = y.limit(x = 0, dir='+')
ll = y.limit(x = 0, dir='-')
if (lr == ll and lr != +oo and lr != -oo):
    print("точка разрыва первого рода")
if (ll == +oo and lr == +oo or ll == -oo and lr == -oo):
    print("точка разрыва второго рода")
```

<h2>Асимптоты</h2>
Вертикальные асимптоты проходят через точки разрыва (когда предел функции в этой точке стремится к бесконечности)<br>
Горизонтальные асимптоты проходят через точку с координатой х, которая является пределом функции при устремлении аргумента функции к бесконечности<br>
Наклонные асимптоты задаются графиком прямой $kx+b$, где за наклон будет отвечать предел отношения функции к аргументу, где поледний стремится к бесконечности. За смещение отвечает предел функции $f(x)-kx$, где х стремится к бесконечности<br>

```python
asimpt = plot(y, -3, 3)
if (y.limit(x = 0) == oo):
    asimpt += plot(0, linestyle="dashed", color="red")
    show(html("<p style='color:#FF0000'>Есть вертикальная асиптота</p>"))
    
if (y.limit(x = oo) != oo):
    asimpt += plot(y.limit(x = oo), linestyle="dashed", color="green")
    show(html("<p style='color:#00FF00'>Есть горизонтальная асиптота</p>"))

f1 = y(x)/x
k = f1.limit(x = oo)
f2 = y(x)-k*x
b = f2.limit(x = oo)
if (k != oo and b != oo):
    asimpt += plot(k*x+b, linestyle="dashed", color="#9ACD32")
    show(html("<p style='color:#9ACD32'>Есть наклонная асиптота</p>"))

show(asimpt)
```
